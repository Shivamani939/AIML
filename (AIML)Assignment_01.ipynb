```python
# 3. Mathematical Operations

import numpy as np

arr1 = np.array((1,2,3,4,5))
arr2 = np.array((5,4,3,2,1))

# multiplication

multi = arr1*arr2
print("Multiplication ",multi)

# Division

div = arr1/arr2
print("Division ",div)

# PowerOff

power = arr1**arr2
print("PowerOff ",power)
```

```python
# 4. Combining text with numpy function
textual_output = f"Addition of Two: {arr1} + {arr2} = {arr1 + arr2}"
print(textual_output)
```

```python
# 5. Using numpy functions
sin_values = np.sin(arr1)
log_values = np.log(arr1)
log2_values = np.log2(arr1)
exp_values = np.exp(arr1)

# Displaying the results
print("Sin values of array1:", sin_values)
print("Log values of array1:", log_values)
print("Log2 values of array1:", log2_values)
print("Exp values of array1:", exp_values)
```

# **Part 2 – Visualization of Data using matplotlib, pyplots Packages**

```python
# 1.

import matplotlib.pyplot as plt

people = ['kiran','arun','vijay','varun']
age =[25, 30, 35, 40]
height =[145, 151, 165, 173]
weight=[45, 55, 65, 75]

plt.scatter(age,weight, c="blue")
plt.title("Age VS Weight")
plt.xlabel("Age")
plt.ylabel("Weight")

plt.show()



```

**Justification**: The scatter plot helps visualize the correlation between age and weight. We expect to see if weight increases with age among the given data.

```python
# 3. Bar Chart

plt.figure(figsize = (6,4))
plt.bar(people,height,color="red")

plt.title("Names and Height of The People")
plt.xlabel("Names")
plt.ylabel("Height")
plt.show()
```

**Justification**: The bar chart provides a clear visual comparison of the heights of the individuals in the dataset.

```python
# 3. Histogram

plt.figure(figsize = (6,4))
plt.hist(weight,color='blue',edgecolor='black')

plt.title("Weight Distribution")
plt.xlabel("Weight")
plt.ylabel("Frequency")

plt.show()
```

**Justification**: The histogram shows the frequency distribution of weights. It helps identify the most common weight range among the individuals.

**Explanation :**

**> Scatter Plot** (Age vs Weight): This plot shows the relationship between age and weight. From the scatter plot, we can observe if there's any trend or correlation, such as weight increasing with age.

**> Bar Chart** (Height of Individuals):
This bar chart provides a clear comparison of each individual's height, allowing us to see who is the tallest and who is the shortest at a glance.

**> Histogram** (Weight Distribution): The histogram displays the distribution of weights among the individuals, showing how the weights are spread out and the frequency of each weight range.
